```r
library(fBasics)
```

```r
da = read.table('ch1data/d-mmm-0111.txt',header=T)
```

```r
mmm = da[,2]
t(basicStats(mmm))
```

```r
t.test(mmm)
# with p-value 0.35, which is greater than 0.05. Thus, the null hypothesis of zero mean cannot be rejected at the 5% level.
```

```r
normalTest(mmm,method='jb')
# the normality assumption for the daily 3M simple returns is rejected.
```

```r
hist(mmm,nclass=30) # Histogram
```

```r
x=seq(-.1,.1,.001)
y1=dnorm(x,mean(mmm),stdev(mmm))
d1=density(mmm)  # Obtain density estimate
plot(d1$x,d1$y,xlab='x',ylab='y',type='l')
lines(x,y1,lty=2)
```

```r
library(quantmod)
getSymbols("AAPL",from="2015-01-03",to="2020-01-03")
X=AAPL[,1:4] # Locate open, high, low, and close prices
xx=cbind(as.numeric(X[,1]),as.numeric(X[,2]),as.numeric(X[,3]),as.numeric(X[,4]))

source("ohlc.R") # Compile the R script
ohlc(xx,xl="days",yl="price",title="Apple Stock")

source("ma.R")  # Compile R script
x1=as.numeric(AAPL$AAPL.Close) # Locate close price
ma(x1,21)
```

```r
da=read.table("ch1data/m-ibmsp-2611.txt",header=T)
ibm=log(da$ibm+1) # Transform to log returns
sp=log(da$sp+1)
tdx=c(1:nrow(da))/12+1926 # Create time index
```

```r
par(mfcol=c(2,1))
plot(tdx,ibm,xlab='year',ylab='lrtn',type='l')
title(main='(a) IBM returns')
plot(tdx,sp,xlab='year',ylab='lrtn',type='l') # X-axis first.
title(main='(b) SP index')
```

```r
cor(ibm,sp)  # Obtain sample correlation
```

```r
m1=lm(ibm~sp)  # Fit the Market Model (linear model)
summary(m1)
```

```r
plot(sp,ibm,type='p',cex=0.8)  # Obtain scatter plot
abline(a=0.007768,b=0.806685) # Add the linear regression line
```

```r
rt=cbind(ibm,sp)
apply(rt,2,mean)
```
